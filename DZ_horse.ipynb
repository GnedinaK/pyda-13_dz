### Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты.    
Принять и обосновать решение о дальнейшей работе с ними.

```python
# Выбросы в показателях температуры

q1 = data_horse.rectal_temperature.quantile(0.25)
q3 = data_horse.rectal_temperature.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = data_horse[data_horse.rectal_temperature.between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal_temperature')
remove_outliers
```

```python
print(remove_outliers.rectal_temperature.mean())
print(remove_outliers.rectal_temperature.median())
print(data_horse.rectal_temperature.mean() - remove_outliers.rectal_temperature.mean())
print(data_horse.rectal_temperature.median() - remove_outliers.rectal_temperature.median())
```

```python
# В показателях температуры отклонения незначительны
data_horse['rectal_temperature'].describe()
```

```python
# Выбросы в показателях пульса

q1_p = data_horse.pulse.quantile(0.25)
q3_p = data_horse.pulse.quantile(0.75)
iqr_p = q3_p - q1_p
lower_bound_p = q1_p - (1.5 * iqr_p) 
upper_bound_p = q3_p + (1.5 * iqr_p)
remove_outliers_p = data_horse[data_horse.pulse.between(lower_bound_p, upper_bound_p, inclusive=True)].sort_values('pulse')
remove_outliers_p
```

```python
print(remove_outliers_p.pulse.mean())
print(remove_outliers_p.pulse.median())
print(data_horse.pulse.mean() - remove_outliers_p.pulse.mean())
print(data_horse.pulse.median() - remove_outliers_p.pulse.median())
```

```python
data_horse['pulse'].describe()
```

*Поскольку выбросы не повлияли значительным образом на показатели среднего и медианы, их исключение нецелесообразно. Продолжим работать с представленными значениями, полагая, что они имеют объективную природу, связанную с показателями степени тяжести заболевания лошадей.*

### Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов.    
Принять и обосновать решение о методе работы с пропусками по каждому столбцу,    
сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
for col in data_horse.columns:
    pct_missing = data_horse[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
# не знаю, что я здесь хотела увидеть

import matplotlib.pyplot as plt

for col in data_horse.columns:
    missing = data_horse[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  
        print('created missing indicator for: {}'.format(col))
        data_horse['{}_ismissing'.format(col)] = missing


# на основе индикатора строим гистограмму
ismissing_cols = [col for col in data_horse.columns if 'ismissing' in col]
data_horse['num_missing'] = data_horse[ismissing_cols].sum(axis=1)

data_horse['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')
```
