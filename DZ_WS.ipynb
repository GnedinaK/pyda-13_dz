```python
KEYWORDS = ['python', 'парсинг', 'принцип', 'Radio', 'scrum']
URL = 'https://habr.com/ru/all/'
req = requests.get(URL)
soup = BeautifulSoup(req.text, 'html.parser')
preview = soup.find_all('article', class_ = 'post post_preview' )

post_times = []
post_titles = []
post_links = []


for post in preview:
    post_time = post.find('span', class_ = 'post__time').text
    post_title = post.find('a', class_= 'post__title_link').text
    post_text = post.find('div',class_ = 'post__text').text
    post_link = post.find('a', class_= 'post__title_link').get('href')
    for word in KEYWORDS:
        if word.lower() in post_title.lower() or word.lower() in post_text.lower():
            post_times.append(post_time)
            post_links.append(post_link)
            post_titles.append(post_title)
        else:
            pass
    print(post_time)
    print(post_title)
    print(post_link)
```

```python
# или вот так как-то чтоли

KEYWORDS = ['python', 'парсинг', 'принцип', 'Radio', 'scrum', 'Flutter', 'BuildRoot', 'Composition']
URL = 'https://habr.com/ru/all/'

def parse_habr_prewiev(URL, KEYWORDS):
    
    df_parse_habr_prewiev = pd.DataFrame()
    req = requests.get(URL)
    soup = BeautifulSoup(req.text, 'html.parser')
    preview = soup.find_all('article', class_ = 'post post_preview' )


    for post in preview:
        post_title = post.find('a', class_ = 'post__title_link').text
        post_time = post.find('span', class_ = 'post__time').text
        post_text = post.find('div', class_ = 'post__text').text
        post_link = list(map(lambda x: x.find('a', class_='post__title_link').get('href'), preview))
        
        for word in KEYWORDS:
            if word.lower() in post_title.lower() or word.lower() in post_text.lower():
                row = {'Дата': post_time, 'Заголовок': post_title, 'Ссылка': post_link}
                df_parse_habr_prewiev = pd.concat([df_parse_habr_prewiev, pd.DataFrame([row])])
            else:
                pass
    return df_parse_habr_prewiev
            
parse_habr_prewiev(URL, KEYWORDS)
```

```python
# В одной ячейке для наглядности

EMAIL = ['xxx@x.ru', 'yyy@y.com']
url = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
headers = {'Vaar-Version': '0', 'Vaar-Header-App-Product': 'hackcheck-web-avast' }
req = requests.post(url, json = {'emailAddresses': EMAIL}, headers = headers)
breaches = pd.DataFrame(req.json()['breaches']).transpose().reset_index()
data = breaches.filter(items = ['index', 'site', 'description', 'publishDate'])
data_r = data.rename({'index': 'index', 
                      'site': 'источник утечки', 
                      'description': 'описание утечки', 
                      'publishDate': 'дата утечки'}, axis = 1
                     )
edit = data_r[['index', 'дата утечки', 'источник утечки', 'описание утечки']]
emails = pd.DataFrame(req.json()['summary'])
emailAdress = []
emailID = []

for email, index in emails.items():
    for id_ in index['breaches']:
        emailAdress.append(email)
        emailID.append(id_)

        
summary = pd.DataFrame()
summary['email'] = emailAdress
summary['index'] = emailID
summary['index'] = summary['index'].astype(str)
emails_Hack_Ckeck = edit.merge(summary, on = 'index', how = 'left')
emails_Hack_Ckeck[['email', 'дата утечки', 'источник утечки', 'описание утечки']] 
```
